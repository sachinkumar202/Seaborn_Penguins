## Line Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
var=[1,2,3,4,5,6,7,8]
var_1= [3,5,7,9,99,77,66,55]
df=pd.DataFrame({'var':var,'var_1':var_1})
```

```python
plt.plot(var,var_1)
plt.show()
```

```python
sns.lineplot(x='var',y='var_1',data=df)
plt.show()
```

```python
sns.lineplot(x=var,y=var_1)
plt.show()
```

# Countplot

```python
data=sns.load_dataset('tips')
data.head(20)
```

```python
sns.countplot(x='sex',data=data,hue='smoker',color='c')
plt.show()
```

```python
sns.countplot(y='sex',data=data,hue='smoker')
plt.show()
```

```python
sns.barplot(x='sex',y='size',data=data)
plt.show()
```

# Violin plot

```python
sns.violinplot(x='day',y='total_bill',data=data,linewidth=3,palette='cool_r')
plt.show()
```

```python
sns.violinplot(x='day',y='total_bill',data=data,hue='time')
plt.show()
```

# Pair plot

```python
sns.pairplot(data,hue='sex',kind='kde',)
plt.show()
```

## Strip plot

```python
sns.stripplot(x='day',y='total_bill',data=data,hue='sex')
```

## Box plot

```python
sns.boxplot(x='day',y='total_bill',data=data,hue='sex')
plt.show()
```

## Factor plot

```python
sns.factorplot(x='size',y='tip',data=data,hue='sex')
plt.show()
```

## catplot

```python
sns.catplot(x='tip',y='size',data=data,hue='sex',kind='bar')
plt.show()
```

# Styling plot

```python
sns.set_style('darkgrid')
sns.barplot(x='day',y='total_bill',data=data)
plt.grid()
plt.show()
```

```python
# remove axis line
sns.set_style('white')
sns.barplot(x='day',y='total_bill',data=data)

sns.despine()
plt.show()
```

## Scale and Context

```python
sns.set_style('darkgrid')

plt.figure(figsize=(12,4))

sns.barplot(x='day',y='total_bill',data=data)

plt.grid()
plt.show()
```

```python
sns.set_style('darkgrid')
sns.set_context('poster',font_scale=0.52)
sns.barplot(x='day',y='total_bill',data=data,palette='cool')
plt.grid()
plt.show()
```

# facet-grid

```python
fg=sns.FacetGrid(data,col='sex',hue='day',palette='cool')
fg.map(plt.bar,'total_bill','tip').add_legend()

plt.show()
```
